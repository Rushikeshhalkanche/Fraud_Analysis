# **Data Vizualization**

```python
# lets have look at our target variable's counts
isFraud_values=df['isFraud'].value_counts()
isFraud_values
```

```python
# Visualizing the Target variable's class distribution.
labels = ["Not Fraud", "Fraud"] # Changed the labels to match the actual values in the isFraud variable
isFraud_values.plot.pie(explode=[0.05]*2,labels=labels,autopct='%1.1f%%',figsize=(12,8),fontsize=15) # Changed the explode list to have two values, one for each slice
```

```python
# Chart - 3 visualization code
# Assuming data_num should contain numerical features from df
data_num = df.select_dtypes(include=['number']) # Selects columns with numerical data types

for i in data_num.columns:
    plt.hist(df[i],color='#808000')
    plt.title(i)
    plt.show()
```

```python
# Chart - 5 visualization code
# dealing with outliers values
for i in data_num.columns:
    sns.boxplot(data_num[i],color='#FFA07A')
    plt.title(i)
    plt.show()
```
